```python
%matplotlib inline
import pandas as pd
```

```python
cast = pd.read_csv('data/cast.csv')
cast.head()
```

```python
release_dates = pd.read_csv('data/release_dates.csv', index_col=None,
                                      parse_dates=['date'], infer_datetime_format=True)
release_dates.head()
```

### Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

```python
christmas_usa = release_dates[(release_dates.title.str.contains('Christmas')) & (release_dates.country == 'USA')]
christmas_usa.date.dt.month.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the months in which movies whose titles start with "The Hobbit" are released in the USA.

```python
hobbit_usa = release_dates[(release_dates.title.str.contains('The Hobbit')) & (release_dates.country == 'USA')]
hobbit_usa.date.dt.month.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Romance" in their title tend to be released in the USA.

```python
hobbit_usa = release_dates[(release_dates.title.str.contains('Romance')) & (release_dates.country == 'USA')]
hobbit_usa.date.dt.dayofweek.value_counts().sort_index().plot(kind='bar')
```

### Make a bar plot of the day of the week on which movies with "Action" in their title tend to be released in the USA.
