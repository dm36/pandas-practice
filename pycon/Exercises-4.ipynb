### Define a year as a "Superman year" whose films feature more Superman characters than Batman. How many years in film history have been Superman years?

```python
splusb = cast[(cast.character == 'Superman') | (cast.character == 'Batman')]
splusb
```

```python
splusb = splusb.groupby(['year', 'character']).size()
splusb
```

```python
# The values for character above- Batman and Superman now become the columns
splusb = splusb.unstack('character')
```

```python
splusb
```

```python
# Fills in missing data with 0's so you can take differences
splusb = splusb.fillna(0)
```

```python
splusb['difference'] = splusb.Superman - splusb.Batman
```

```python
len(splusb[splusb.difference > 0])
```

### How many years have been "Batman years", with more Batman characters than Superman characters?

```python
len(splusb[splusb.difference < 0])
```

### Plot the number of actor roles each year and the number of actress roles each year over the history of film.

```python
actor = cast[cast.type == 'actor']
actor.groupby('year').size().sort_index().plot()
```

```python
actress = cast[cast.type == 'actress']
actress.groupby('year').size().sort_index().plot()
```

```python
year_type = cast.groupby(['year', 'type']).size()
year_type
```

```python
unstacked = year_type.unstack('type')
unstacked
```

```python
unstacked.plot()
```

### Plot the number of actor roles each year and the number of actress roles each year, but this time as a kind='area' plot.

```python
unstacked.plot(kind='area')
```

### Plot the difference between the number of actor roles each year and the number of actress roles each year over the history of film.

```python
year_type = cast.groupby(['year', 'type']).size()
year_type = year_type.unstack('type')
year_type['difference'] = year_type.actor - year_type.actress
year_type
year_type.difference.sort_index().plot()
```

### Plot the fraction of roles that have been 'actor' roles each year in the history of film.

```python
year_type = cast.groupby(['year', 'type']).size()
year_type = year_type.unstack('type')
year_type['fraction'] = (year_type.actor) / (year_type.actor + year_type.actress)
year_type.fraction.plot(ylim=[0,1])
```

### Plot the fraction of supporting (n=2) roles that have been 'actor' roles each year in the history of film.

```python
cast_n_2 = cast[cast.n == 2]
year_type = cast_n_2.groupby(['year', 'type']).size()
year_type = year_type.unstack('type')
year_type['fraction'] = (year_type.actor) / (year_type.actor + year_type.actress)
year_type.fraction.plot(ylim=[0,1])
```

### Build a plot with a line for each rank n=1 through n=3, where the line shows what fraction of that rank's roles were 'actor' roles for each year in the history of film.

```python
cast_n_through_3 = cast[(cast.n >= 1) & (cast.n <= 3)]
year_type = cast_n_through_3.groupby(['n', 'year', 'type']).size()
year_type = year_type.unstack('type')
year_type['fraction'] = (year_type.actor) / (year_type.actor + year_type.actress)
year_type = year_type.drop(columns=['actor', 'actress'])
year_type = year_type.unstack('n')
year_type
year_type.plot(ylim=[0,1])
```
